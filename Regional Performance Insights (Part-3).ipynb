```python
final_orders_data = pd.merge(orders_list_data,orders_data,on=['Order ID'],how="inner")
```

```python
final_orders_data
```

```python
final_orders_data.shape
```

```python
State_wise_order_list1 = final_orders_data.groupby('State')['Order ID'].count().reset_index()
State_wise_order_list1
```

```python
State_wise_order_list = final_orders_data.groupby('State')['Order ID'].count().nlargest(5).reset_index()
State_wise_order_list.columns = ['State','Order Counts']
```

```python
State_wise_order_list
```

```python
plt.figure(figsize=(12,5))
sns.barplot(x="State",y="Order Counts",data=State_wise_order_list,palette = "icefire")
plt.title("Top 5 states order Counts")
plt.xticks(rotation=45)
plt.show()
```

```python
top_states = State_wise_order_list["State"]
top_states
```

```python
filtered_data = final_orders_data[final_orders_data['State'].isin(top_states)]
filtered_data
```

```python
filtered_data['State'].unique()
```

```python
sales_profit_summary = (
    filtered_data.groupby("State").agg({"Order ID":"count","Amount": "sum", "Profit": "mean"}).reset_index()
)
```

```python
sales_profit_summary = sales_profit_summary.set_index("State").loc[top_states].reset_index()
sales_profit_summary
```

```python
sales_profit_summary.columns = ["State","Order_count","Total_sales","Average_Profit"]
sales_profit_summary
```

```python
filtered_data['City'].unique()
```

```python
filtered_data
```

```python
city_sales_data = filtered_data.groupby(['State','City']).agg({"Order ID":"count","Amount": "sum", "Profit": "mean"})
```

```python
city_sales_data.columns = ["Order Counts","Total Sales","Average Profit"]
```

```python
city_sales_data
```

```python
city_sales_profit = city_sales_data.sort_values(by=["State", "Total Sales"], ascending=[True, False])
```

```python
city_sales_profit
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=city_sales_profit, x="Total Sales", y="City", hue="State", dodge=False, palette="viridis")
plt.xlabel("Total Sales")
plt.ylabel("City")
plt.title("Total Sales per City")
plt.legend(title="State", bbox_to_anchor=(1, 1))
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=city_sales_profit, x="Average Profit", y="City", hue="State", dodge=False, palette="viridis")
plt.xlabel("Average Profit")
plt.ylabel("City")
plt.title("Average Profit per City")
plt.legend(title="State", bbox_to_anchor=(1, 1))
plt.show()
```
